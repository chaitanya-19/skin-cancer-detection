```python
import cv2
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("E:\df_v4.csv",index_col=False)
df=df.drop(["Unnamed: 0"],axis=1)
df
```

```python
def augment(img):
    img1 = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)
    img2 = cv2.rotate(img, cv2.ROTATE_90_COUNTERCLOCKWISE)
    img3 = cv2.rotate(img, cv2.ROTATE_180)
    a=[img1,img2,img3]
    img_ran1=random.sample(a,1) 
    img_ran1=np.array(img_ran1).reshape(80,80)
   # plt.imshow(img_ran1)
    img4 = cv2.flip(img_ran1, 0)
    img5 = cv2.flip(img_ran1, 1)
    a=[img4,img5]
    img_ran2=random.sample(a,1)
    img_ran2=np.array(img_ran2).reshape(80,80)
    a=[img_ran1,img_ran2]
    
    return random.sample(a,1)
```

```python
df_x=df.drop(["target"],axis=1)
df_y=np.array(df["target"])
arr=[]
d={}
num=[]
for i in range(9590):
    num.append(i)
    temp=df_x.iloc[i,:]
    arr.append(np.array(temp).reshape(80,80))
    d[i]=df_y[i]
```

```python
d
```

```python
test_x_augmented=[]
test_y_augmented=[]
for i in range(1000):
    n=1
    index=random.sample(num,n)
    
    test_x_augmented.append(augment(arr[int(index[0])]))
    test_y_augmented.append(d[int(index[0])])
    
print(test_x_augmented)  
```

```python
pd.Series(test_y_augmented).value_counts()
```

```python
data_flat=[]
for i in range(len(test_x_augmented)):
    temp=np.array(test_x_augmented[i]).flatten()
    data_flat.append(np.array(temp))
data_flat
```

```python
df=pd.DataFrame(data_flat)
df
```
